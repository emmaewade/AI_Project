```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import statsmodels.api as sm  
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA

%matplotlib inline

data = pd.read_csv('train_covid_data.csv', index_col = 'date', parse_dates = True) 
data.head()
```

```python
data.isnull().values.any()
```

```python
# plot new cases over all years
plt.figure(figsize=(20,4))
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Confirmed Cases')
plt.plot(data.index, data['new_cases']);     
```

```python
# plot Confirmed Deaths over all years
plt.figure(figsize=(20,4))
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Confirmed Deaths')
plt.plot(data.index, data['new_deaths']);   
```

```python
# let's take a closer look: March 2020 to December 2021
start = datetime.datetime(2020,3,1)
end = datetime.datetime(2021,12,1)

vaccine_data = data.loc[(data['iso_code'] == 'USA')]

plt.figure(figsize=(20,4))
ax = plt.gca()
ax.set_xlim(start, end)
ax.set_ylim(0, 300000)
ax.set_ylabel('Confirmed Cases')
plt.scatter(vaccine_data.index, vaccine_data['new_cases'])     
ax.legend(['USA DATA'])
```

```python
# plot outliers by year
data_boxplot = []
for year in range(2020, 2023):
    data_boxplot.append(vaccine_data.loc[(vaccine_data.index.year == year)]['new_cases'])
    
import seaborn as sns
plt.figure(figsize=(8,5))
ax = plt.gca()
sns.boxplot(data=data_boxplot, color='w', whis=1.5)
ax.set_xlabel('Year')
ax.set_ylabel('Confirmed Cases for USA');
# plt.boxplot(data_boxplot, whis=1.5, showfliers=True);   
# use seaborn over matplotlib since matplotlib is not showing outliers
```

```python
def test_stationarity(ts):
    stats = ['Test Statistic','p-value','Lags','Observations']
    df_test = adfuller(ts, autolag='AIC')
    df_results = pd.Series(df_test[0:4], index=stats)
    for key,value in df_test[4].items():
        df_results['Critical Value (%s)'%key] = value
    print(df_results)
test_stationarity(vaccine_data['new_cases'])
```

```python
vaccine_data['diff'] = vaccine_data['new_cases'] - vaccine_data['new_cases'].shift(1)  
test_stationarity(vaccine_data['diff'].dropna(inplace=False))
```

```python
# ACF, PACF graphs to help determine order of ARIMA model, again statsmodel has these handy functions built-in
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(vaccine_data['new_cases'][1:], lags=60, ax=ax1)   # first value of diff is NaN
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(vaccine_data['new_cases'][1:], lags=60, ax=ax2)
```

```python
vaccine_data['new_cases.float'] = vaccine_data['new_cases'].astype(float)
print (type(vaccine_data['new_cases.float'][5]))
```

```python
# fit ARIMA model
fig = plt.figure(figsize=(20,8))
model = ARIMA(vaccine_data['new_cases.float'], order=(1,0,0)) 
ax = plt.gca()
results = model.fit() 
plt.plot(vaccine_data['new_cases'])
plt.plot(results.fittedvalues, color='red')
ax.legend(['New Cases', 'Forecast'])

print(results.summary())
```

```python
# fit ARIMA model
fig = plt.figure(figsize=(20,8))
model = ARIMA(vaccine_data['new_cases.float'], order=(1,0,1)) 
ax = plt.gca()
results = model.fit() 
plt.plot(vaccine_data['new_cases'])
plt.plot(results.fittedvalues, color='red')
ax.legend(['New Cases', 'Forecast'])

print(results.summary())
```

```python
# forecast quick and dirty
# TODO: clean this up
fig = plt.figure(figsize=(20,8))
num_points = len(vaccine_data['new_cases'])
x = results.predict(start=(0), end=(1000), dynamic=False)

plt.plot(vaccine_data['new_cases'][:750])
plt.plot(x, color='r')
```
